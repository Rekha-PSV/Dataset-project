**Step 5 : Data visualization**

```python
import matplotlib.pyplot as plt
```

```python
data.columns
```

**Harzontal Bar Graph**

```python
# Assuming 'data' is your DataFrame
grouped = data.groupby('PRI_FDA Industry Name')['PRI_FDA Industry Code'].mean()
grouped.plot(kind='barh')
plt.title('PRI_FDA Industry Code by PRI_FDA Industry Name')
plt.xlabel('PRI_FDA Industry Name')
plt.ylabel('PRI_FDA Industry Code')
plt.show()
```

**pie chart**

```python
# Assuming 'data' is your DataFrame
grouped = data.groupby('PRI_Product Role').size()

# Plotting the pie chart
grouped.plot(kind='pie', autopct='%1.1f%%', startangle=90)
plt.title('CI_Gender')
plt.ylabel('')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

**Line chart**

```python
# Assuming 'data' is your DataFrame
grouped = data.groupby('RA_CAERS Created Date')['PRI_FDA Industry Code'].mean()
grouped.plot(kind='line', marker='o')
plt.title('PRI_FDA Industry Code by RA_CAERS Created Date')
plt.xlabel('RA_CAERS Created Date')
plt.ylabel('Mean PRI_FDA Industry Code')
plt.xticks(rotation=45)
plt.show()
```

**Scatter Graph**

```python
 # Assuming 'data' is your DataFrame
grouped = data.groupby('CI_Age Unit')
for name, group in grouped:
    plt.scatter(group['CI_Gender'], group['PRI_FDA Industry Code'], label=name, alpha=0.5)
plt.title('PRI_FDA Industry Code vs CI_Gender by CI_Age Unit')
plt.xlabel('PRI_FDA Industry code')
plt.ylabel('CI_Gender')
plt.legend(title='CI_Age Unit')
plt.show()
```

**Bar Graph**

```python
# Assuming 'data' is your DataFrame
grouped = data.groupby('PRI_FDA Industry Code')['CI_Age at Adverse Event'].mean()
grouped.plot(kind='bar')
plt.title('Mean CI_Age at Adverse Event by PRI_FDA Industry Code')
plt.xlabel('PRI_FDA Industry Code')
plt.ylabel('Mean CI_Age at Adverse Event')
plt.show()
```

**Step 6 : Aggregate Function**

```python
# groupby function for Customer ID and Quantity
df = pd.DataFrame(data)
grouped=df.groupby('PRI_FDA Industry Code')['CI_Age at Adverse Event'].count()
print(grouped)
```

```python
# Assuming data is your DataFrame
aggregate_functions = {
    'PRI_FDA Industry Code': ['sum', 'mean', 'median', 'min', 'max', 'std', 'count'],
}
summary_statistics = data.agg(aggregate_functions)
print(summary_statistics)
```
